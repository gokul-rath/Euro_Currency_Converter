```python
# Input Rapidapikey provided by the rapidapi webpage
RAPIDAPIKEY = str(input("Enter your RAPIDAPIKEY to proceed:"))
```

```python
# import
import http.client  
import json 
import pandas as pd
import numpy as np
```

```python
import veroviz as vrv
vrv.checkVersion()
```

```python
# Establish the connection to the particular API 
conn = http.client.HTTPSConnection("currency-converter5.p.rapidapi.com")

headers = {
    'x-rapidapi-host': "currency-converter5.p.rapidapi.com",
    'x-rapidapi-key': RAPIDAPIKEY
    }
```

```python
# First we will find the List of available currencies:
# The select API have been provided free of cost and is available for everyone
conn.request("GET", "/currency/list", headers=headers)

# Get the data response and desired format text
resp  = conn.getresponse()
data = resp.read().decode("utf-8")

#obtain a dictionary so we can display all the most important currencies exchanges comparing to euro
print(type(data))
print(data)
currency = json.loads(data)
```

```python
# Visualize the full denomination for each currency.
currency 
```

```python
# The keys used in the list of currencias are : the name of the currency and if the actual currency is been used in the market:
currency.keys()
```

```python
# Inside the key of currencias we can see the meaning for each one
for key in currency.keys():
    print("key =", key)
    print(currency[key])
    print("\n")
```

```python
# changed to a list of currencies from the dictionary
currency['currencies']
```

```python
# Next, we will check the currency exchange rate based on 1 euro
#This means that we will obtain all the currencies values equal to 1 euro 
conn.request("GET", "/currency/convert", headers=headers)

resp = conn.getresponse()
data = resp.read().decode("utf-8")
convert = json.loads(data)
convert
```

```python
# The currencies value are obtained from a constantly updated finance webpage.
for key in convert.keys():
    print("key =", key)
    print(convert[key])
    print("\n")
```

```python
# change to list visualization
convert['rates']
```

```python
# panda dataframe convert rate 
convert_df= pd.DataFrame(convert['rates'])
convert_df
```

```python
# get the Historical Currency Rates specifying the year - month - day
conn.request("GET", "/currency/historical/2019-12-10", headers=headers)

resp = conn.getresponse()
data = resp.read().decode("utf-8")
historical = json.loads(data)
historical
```

```python
# we can make an analisys of the exchange rate of euros and dollars trhoughout the pre corona dates and post corona dates 
convert_df= pd.DataFrame(historical['rates'])
convert_df
```
